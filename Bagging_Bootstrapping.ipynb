# **Baging and Bootstrapping learning**

## **0. Imports**

```python
from sklearn.ensemble import BaggingClassifier, BaggingRegressor
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

## **Build a Bagging Regressor**

```python
np.random.seed(31415)
bostonds= load_boston()
```

```python
X_train,X_test, y_train,y_test = train_test_split(bostonds.data, bostonds.target, test_size=0.3)
```

```python
# Aggregate results from 15 linear regression models
bagging_regressor = BaggingRegressor(
    base_estimator = LinearRegression(),
    n_estimators= 100
)
```

```python
bagging_regressor.fit(X_train,y_train)
test_predictions = bagging_regressor.predict(X_test)
test_predictions
```

```python
t = np.arange(0.0, y_test.shape[0],1)
```

```python
fig, ax = plt.subplots()
ax.plot(t, y_test)
plt.show()
```

```python
ax.set(xlabel='Observation', ylabel='Price', title='Housing Princes')
ax.plot(t, test_predictions)
ax.grid()
plt.show()
```

```python
r2 = r2_score(test_predictions, y_test)
print('R2 score of model is {:.3f}'.format(r2))
```
