## 1. Impulse Sampling
image.png


<h3>Spectrum of ideal sampled signal is given by: G(f) = fs .[ Σ X(f-n fs)] </h3>


```python
# Amplitude & Frequency
a = 5
f= 2

#Generate train of impulses, Sine wave and impulse sampling

t = np.arange(0,2,0.01)
x1 =[]

for i in range(len(t)):
  x1.append(1)

x2 = a * np.sin(2 * np.pi * f * t)
y = x1 * x2

```

## 1.1 Impluse Signal: Amplotude Time

```python
#Plotting 
plt.stem(t, x1)
plt.title('Impulse Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## 1.2 Sine Wave

```python
plt.plot(t, x2)
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude ')
plt.grid()
plt.show()
```

## 1.3 Plot Signal Sampled implulse and sine

```python
plt.stem(t, y)
plt.title('PAM Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```
