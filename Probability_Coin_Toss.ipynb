## Imports


```python
import numpy as np
from matplotlib import pyplot as plt
```

## Probablity Practice Fuse 

```python
#coin Example

coin = ['H','T']

# list to save vals of toss 
freq_list = []

# number of head counts during tosses
no_of_heads = 0


#simulate tosses
for flip in range(1000):
  toss_result = np.random.choice(coin) #this will take H or T
  no_of_tosses = flip +1

  if toss_result == 'H':
    no_of_heads += 1
    f = no_of_heads / no_of_tosses
    freq_list.append(f)
  else:
    f = no_of_heads / no_of_tosses
    freq_list.append(f)
```

## Plot all frequencies



```python
plt.figure(figsize=(15,5))
plt.ylim((0,1.1))
plt.xlabel('Number of coin tosses')
plt.ylabel('Relative Frequencies')

#plot the Expected Probability
plt.plot([0.5] * len(freq_list), color='red')

#plot the frequency values from experiment
plt.plot(freq_list)

plt.legend(['Actual Probability of heads','Relative Frequencies'])
plt.grid()
plt.show()
```
